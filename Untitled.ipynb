```python
#Question 4

#Classifier model to predict the rating category
```

```python
def func(model, X_train = X_train, X_test = X_test, y_train = y_train, y_test = y_test):
    model.fit(X_train, y_train)
    y_pred_train = model.predict(X_train)
    print("\nResults on the train set")
    print("\nConfusion Matrix\n")
    print(confusion_matrix(y_train, y_pred_train))
    print()
    print("Accuracy score on the train set:\t",accuracy_score(y_train, y_pred_train))
    print("Recall score on the train set:\t",recall_score(y_train, y_pred_train,average="macro" ))
    print("Precision score on the train set:\t",precision_score(y_train, y_pred_train, average="macro"))
    print("Specificity score on the train set:\t",recall_score(y_train, y_pred_train, average="macro"))
    print("f1_score on the train set:\t",f1_score(y_train, y_pred_train,average="macro"))
    print()
    
    y_pred = model.predict(X_test)
    print("\nResults on the test set")
    print("\nConfusion Matrix\n")
    print(confusion_matrix(y_test, y_pred))
    print()
    print("Accuracy score on the test set:\t",accuracy_score(y_pred,y_test))
    print("Recall score on the test set:\t",recall_score(y_test,y_pred,average="macro" ))
    print("Precision score on the test set:\t",precision_score(y_test,y_pred, average="macro"))
    print("Specificity score on the test set:\t",recall_score(y_test, y_pred, average="macro"))
    print("f1_score on the test set:\t",f1_score(y_test, y_pred, average="macro"))
```
